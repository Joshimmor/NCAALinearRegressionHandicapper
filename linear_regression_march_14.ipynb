```python
import numpy as np
import pandas as pd
```

```python
team_data = pd.read_csv("four_factors.csv",index_col=[0])
game_data = pd.read_csv('2022_Season.csv', index_col =[0])
print(team_data.head(),game_data.head())
```

```python
team_data.columns.str.match("Unnamed")
team_data = team_data.loc[:,~team_data.columns.str.match("Unnamed")]
```

```python
team_data = team_data.drop(['Off-OR%.Rank','Off-TO%.Rank','Off-eFG%.Rank','AdjTempo.Rank','AdjOE.Rank','Def-OR%.Rank','Def-TO%.Rank','Def-FTRate.Rank','Def-eFG%.Rank','AdjDE.Rank'],axis=1)
print(team_data.head())
```

```python
feb_game = game_data.iloc[598:1599]
print(feb_game.head())
```

```python
home_team_fg = []
home_team_or = []
home_team_ft = []
home_team_to = []

away_team_fg = []
away_team_or = []
away_team_ft = []
away_team_to = []

point_spread = []
for index,game in feb_game.iterrows():
    away_ID = game['AwayTeamID']
    home_ID = game['HomeTeamID']
    home_team = team_data.loc[team_data['NCAA'] == home_ID]
    away_team = team_data.loc[team_data['NCAA'] == away_ID]
    if home_team.empty or away_team.empty:
        continue 
    home_team_fg.append(float(home_team['Off-eFG%']  -  home_team['Def-eFG%']))
    home_team_or.append(float(home_team['Off-OR%']  -  home_team['Def-OR%']))
    home_team_ft.append(float(home_team['Off-FTRate']  -  home_team['Def-FTRate']))
    home_team_to.append(float(home_team['Off-TO%']  -  home_team['Def-TO%']))
    away_team_fg.append(float(away_team['Off-eFG%']  -  away_team['Def-eFG%']))
    away_team_or.append(float(away_team['Off-OR%']  -  away_team['Def-OR%']))
    away_team_ft.append(float(away_team['Off-FTRate']  -  away_team['Def-FTRate']))
    away_team_to.append(float(away_team['Off-TO%']  -  away_team['Def-TO%']))
    point_spread.append(float(game['HomeTeamScore'] -  game['AwayTeamScore']))
data = {
    'home_team_fg':home_team_fg,
    'home_team_or':home_team_or,
    'home_team_ft':home_team_ft,
    'home_team_to':home_team_to,
    'away_team_fg':away_team_fg,
    'away_team_or':away_team_or,
    'away_team_ft':away_team_ft,
    'away_team_to':away_team_to,
    'point_spread':point_spread
}
cleaned_data = pd.DataFrame(data)
display(cleaned_data)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
print(cleaned_data['point_spread'])
```

```python
# prd_data = cleaned_data.drop('point_spread',axis=1)
# prd = np.array(prd_data['home_team_fg','home_team_or','home_team_ft','home_team_to','away_team_fg','away_team_or','away_team_ft','away_team_to']).reshape(-1,1)
# act = np.array(cleaned_data['point_spread']).reshape(-1, 1)
train_data_dm,test_data_dm = train_test_split(cleaned_data,train_size = 0.8,random_state=3)
independent_var = ['home_team_fg','home_team_or','home_team_ft','home_team_to','away_team_fg','away_team_or','away_team_ft','away_team_to']
regr = LinearRegression()
regr.fit(train_data_dm[independent_var],train_data_dm['point_spread'])
print(regr.score(test_data_dm[independent_var],test_data_dm['point_spread']))
```
