```python
teams_stats = pd.read_csv('four_factors.csv')
# print(teams_stats['Team'])
team_ncaa = pd.read_csv('Teams.csv')
# print(team_ncaa['TeamID'])
IDS = []
to_drop = []
for ts_ind in teams_stats.index:
    for tn_ind in team_ncaa.index:
        if team_ncaa['School'][tn_ind] == teams_stats['Team'][ts_ind]:
            IDS.append(team_ncaa['TeamID'][tn_ind])
            break
        elif tn_ind == len(team_ncaa.index)-1:
            to_drop.append(ts_ind)
# print(len(to_drop)+len(IDS),len(teams_stats.index),len(team_ncaa.index))
teams_stats = teams_stats.drop(labels=to_drop,axis=0)
# print(teams_stats.head())
# print(IDS)
teams_stats['NCAA'] = IDS
teams_stats.to_csv('four_factors.csv')
print(teams_stats.head())
```

```python
# GameSideStats
df = pd.read_csv('2022_season.csv')
home = df['HomeTeamID']
away = df['AwayTeamID']
home_score = df['HomeTeamScore']
away_score = df['AwayTeamScore']
game_spread = []
for i in range(len(home_score)):
    spread = home_score[i] - away_score[i]
    game_spread.append(spread)
game_data = {
    'home_id':home,
    'away_id':away,
    'game_spread':game_spread
}    
game_db = pd.DataFrame(game_data)
game_db.to_csv('training_games_dirty.csv')
```
