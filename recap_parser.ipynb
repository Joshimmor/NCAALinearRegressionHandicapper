```python
from bs4 import BeautifulSoup 
import os 
import codecs
from glob import glob
import re
```

```python
ROOT_DIR = os.path.dirname(os.path.abspath('__file__'))
files = glob('Recaps/*.html')
print(files)
```

```python
class Recaps():
    def __init__(
        self,
        dec_alerts, 
        cp_date, 
        cp_form_id, 
        disponent_owner, 
        registered_owner, 
        charterer, 
        terms, 
        vessel, 
        imo_number, 
        sdwt, 
        draught, 
        loa,
        beam, 
        flag, 
        built, 
        last_cargo, 
        second_last_cargo, 
        third_last_cargo, 
        last_sire, 
        approval_expiry, 
        position_itinerary, 
        now_expected, 
        additional_vessel_info, 
        cargo_description, 
        laydays_commencement, 
        laydays_cancelling, 
        loading_ports, 
        discharging_ports, 
        freight_rate, 
        freight_lump_sum,
        special_provisions, 
        overage, 
        laytime, 
        demurrage, 
        worldscale_tc, 
        freight_payments, 
        commission, 
        other_main_terms):
            self.dec_alerts = dec_alerts 
            self.cp_date = cp_date 
            self.cp_form_id = cp_form_id 
            self.disponent_owner = disponent_owner
            self.registered_owner = registered_owner
            self.charterer = charterer
            self.terms = terms
            self.vessel = vessel 
            self.imo_number = imo_number
            self.sdwt = sdwt 
            self.draught = draught
            self.loa = loa
            self.beam = beam 
            self.flag = flag
            self.built = built
            self.last_cargo = last_cargo
            self.second_last_cargo = second_last_cargo
            self.third_last_cargo = third_last_cargo 
            self.last_sire = last_sire
            self.approval_expiry = approval_expiry 
            self.position_itinerary  = position_itinerary
            self.now_expected = now_expected
            self.additional_vessel_info = additional_vessel_info 
            self.cargo_description = cargo_description 
            self.laydays_commencement = laydays_commencement
            self.laydays_cancelling = laydays_cancelling 
            self.loading_ports= loading_ports
            self.discharging_ports = discharging_ports
            self.freight_rate = freight_rate 
            self.freight_lump_sum = freight_lump_sum
            self.special_provisions = special_provisions 
            self.overage = overage 
            self.laytime = laytime 
            self.demurrage = demurrage 
            self.worldscale_tc = worldscale_tc 
            self.freight_payments = freight_payments 
            self.commission = commission 
            self.other_main_terms = other_main_terms
    def get_info(self):
        print(vars(self))
```
